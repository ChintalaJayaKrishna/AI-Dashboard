```python
df.drop_duplicates(inplace=True)

print("Shape after removing duplicates:")
print(df.shape)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Fare"] = df["Fare"].fillna(df["Fare"].median())
```

```python
total_passengers = len(df)

total_survivors = df["Survived"].sum()

survival_rate = (total_survivors/total_passengers)*100

avg_age = df["Age"].mean()

avg_fare = df["Fare"].mean()

print("Total Passengers:", total_passengers)
print("Total Survivors:", total_survivors)
print("Survival Rate:", round(survival_rate,2),"%")
print("Average Age:", round(avg_age,2))
print("Average Fare:", round(avg_fare,2))
```

```python
gender = input("Enter Gender (male/female/all): ")

if gender.lower() != "all":
    filtered_df = df[df["Sex"] == gender.lower()]
else:
    filtered_df = df

print("Filtered Records:", len(filtered_df))
```

```python
plt.figure(figsize=(6,6))

filtered_df["Survived"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Survival Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.barplot(
    x="Sex",
    y="Survived",
    data=filtered_df
)

plt.title("Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.barplot(
    x="Pclass",
    y="Survived",
    data=filtered_df
)

plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    filtered_df["Age"],
    bins=20,
    kde=True
)

plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    filtered_df["Fare"],
    bins=20,
    kde=True
)

plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Age",
    y="Fare",
    hue="Survived",
    data=filtered_df
)

plt.title("Age vs Fare")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(
    x="Embarked",
    data=filtered_df
)
plt.title("Passengers by Embarkation Port")
plt.show()
```
